Q-9. Given the score of students in multiple exams


Test the hypothesis that the mean scores of all the students are the same. If not, name the
student with the highest score.

```python
import pandas as pd
from scipy.stats import f_oneway

# Create a DataFrame with the given data
data = pd.DataFrame({
    'Name': ['Karan', 'Deepa', 'Karthik', 'Chandan', 'Jeevan'],
    'Exam 1': [85, 70, 90, 75, 95],
    'Exam 2': [90, 80, 85, 70, 92],
    'Final Exam': [92, 85, 88, 75, 96]
})

# Perform one-way ANOVA
f_statistic, p_value = f_oneway(data['Exam 1'], data['Exam 2'], data['Final Exam'])

# Check the p-value
alpha = 0.05  # Set the significance level
if p_value < alpha:
    print("The mean scores of the students are not the same. There is a statistically significant difference.")
else:
    print("The mean scores of the students are the same. There is no statistically significant difference.")
```

The output will indicate whether the mean scores of the students are the same or not based on the p-value. If the p-value is less than the chosen significance level (e.g., 0.05), we reject the null hypothesis and conclude that there is a statistically significant difference in the mean scores.

To find the student with the highest score, you can calculate the total score for each student by summing the scores across all exams and then find the student with the maximum total score:

```python
# Calculate the total score for each student
data['Total Score'] = data['Exam 1'] + data['Exam 2'] + data['Final Exam']

# Find the student with the highest score
student_highest_score = data.loc[data['Total Score'].idxmax(), 'Name']
print("The student with the highest score is:", student_highest_score)
```
